```python
from Bio import SeqIO

original_file = "segregated_protein_Nucleoprotein_sequence.fasta"
run=0

with open(original_file) as original, open("segregated_protein_Nucleoprotein_modified_sequence.fasta", 'w') as corrected:
    records = SeqIO.parse(original_file, 'fasta')
    run=run+1
    print(run)
    for record in records:
        if record.id == 'spike':
            record.id = 'Nucleo'
            record.description = 'protein'
        SeqIO.write(record, corrected, 'fasta')

with open("segregated_protein_Nucleoprotein_modified_sequence.fasta") as corrected:
    records = SeqIO.parse(corrected, 'fasta')
    for record in records:
        print (record.id)             # prints bar
        print (record.name)           # prints bar
        print (record.description)    # prints bar
```
